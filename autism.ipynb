```python
plt.boxplot(x=df['age'])
plt.title('Age outliers')
plt.show()
```

```python
Q1=df['age'].quantile(0.25)
Q3=df['age'].quantile(0.75)
IQR=Q3-Q1
df=df[(df['age']>=1) & (df['age']<=100)]
```

```python
x=df.iloc[:,:-1].values
y=df.iloc[:,-1].values
```

```python
x
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.30,random_state=0)
x_train
```

```python
from sklearn.preprocessing import StandardScaler 
scaler = StandardScaler()
x_train = scaler.fit_transform(x_train)
x_test = scaler.transform(x_test)
```

```python
from sklearn.ensemble import RandomForestClassifier          #DecisionTreeRegression for regression files
model=RandomForestClassifier(n_estimators=100,random_state=42) 
model.fit(x_train,y_train)
y_pred=model.predict(x_test)
y_pred

```

```python
prediction = model.predict([[23,1, 0,0, 0, 0, 0, 1, 0, 1,0]])
prediction
```

```python
import pickle
with open("asd_model.pkl", "wb") as f:
    pickle.dump(model, f)
```

```python
from sklearn.metrics import confusion_matrix,accuracy_score,classification_report
cm=confusion_matrix(y_test,y_pred)
cm
```

```python
score=accuracy_score(y_test,y_pred)*100
score
```

```python
report=classification_report(y_test,y_pred)
report
```
